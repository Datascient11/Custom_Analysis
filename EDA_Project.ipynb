**C - II - 17 Correlation Between PS4 & Xbox One Game Sales**

```python
console_a_game = data[data["Platform"] == 'PS4']["Global_Sales"]
console_b_game = data[data["Platform"] == 'XOne']["Global_Sales"]

t_stat, p_value = stats.ttest_ind(console_a_game, console_b_game)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```
